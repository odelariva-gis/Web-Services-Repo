```python
### Finalized data input for 7/18/2022

list_001 = []

for type_results in type_results_list:
    
    if type_results[0].type not in types_list:
        print("{0} will not be processed.".format(type_results[0].type)) 
        print('\n')
        pass
    
    elif type_results[0].type == 'Scene Service' or type_results[0].type == 'Map Service':
        
        print("{0} will be processed now.".format(type_results[0].type))
        subbing_data = type_results[0].type.replace(' ','').replace("Service","Server")
        print("{0}: subbing data.".format(subbing_data))
        
        for types in type_results:
            list_001_hold = []
            result_url = types.url.replace('rest','admin').replace('/' + str(subbing_data),'.' + str(subbing_data)) + '/iteminfo/manifest/manifest.json'
            print("Getting result for URL: {0}".format(result_url))
            list_001_hold.append(result_url)
            response = requests.post(result_url, params = params, verify = False)
            response_json = response.json()

          
            try:
                print(response_json['resources'][0]['onPremisePath'])
                list_001_hold.append(response_json['resources'][0]['onPremisePath'])
            except:
                print(response_json)
                list_001_hold.append(response_json)        
            
            print('-'*50)
            print('\n')
            list_001.append(list_001_hold)
            pass
        del list_001_hold
            
    elif type_results[0].type == 'Feature Service':

        print("{0} will be processed now.".format(type_results[0].type))
        
    
        for types in type_results:
            list_001_hold = []
            if types.url is None or types.url == '':
                print("NONE TYPE PRESENT IN JSON")
                list_001_hold.append(types.url)
                list_001_hold.append('URL EMPTY OR NOT EXISTING')
                
            else:
                result_url = types.url.replace('rest','admin').replace('/FeatureServer','.FeatureServer') + '/iteminfo/manifest/manifest.json'
                print("Getting result for URL: {0}".format(result_url))
                list_001_hold.append(result_url)
                try:
                    response = requests.post(result_url,params = params, verify = False)
                    response_json = response.json()
                except:
                    list_001_hold.append('JSON CALL FAILED')
                    print("JSON RESPONSE FAILED")
                    print(response_json)
                    pass
                try:
                    split_response_json = response_json['databases'][0]['onServerConnectionString'].split(';')
                    print("DATABASE IN JSON")
                    list_001_hold.append(split_response_json[5])
                    print(split_response_json[5])
                except:
                    print("DATABASE NOT IN JSON")
                    try:
                        result_url = types.url.replace('rest','admin').replace('/FeatureServer','.SceneServer') + '/iteminfo/manifest/manifest.json'
                        response = requests.post(result_url,params = params, verify = False)
                        response_json = response.json()
                        print(response_json['resources'][0]['onPremisePath'])
                        list_001_hold.append(response_json['resources'][0]['onPremisePath'])
                        pass
                    except: 
                        print("RESOURCES NOT IN JSON")
                        try:
                            print(response_json['status'])
                            list_001_hold.append(response_json['status'])
                        except:
                            print('STATUS NOT IN JSON')
                            try:
                                print(response_json['error'])
                                list_001_hold.append(response_json['error'])
                            except:
                                print(response_json)
                                list_001_hold.append(response_json)
                                pass
            
                print("-------------------------------")
                print("\n")
                
            list_001.append(list_001_hold)
        del list_001_hold
print("TOOL DONE!!!")
```
